```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('downloads/books.csv',engine='python')
```

```python
df
```

```python
features=['title','authors','publisher']
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
def combined_features(row):
    return row['title']+" "+row['authors']+" "+row['publisher']
```

```python
df["combined_features"]=df.apply(combined_features,axis=1)
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
cosine_sim
```

```python
Title=df['title'][1]
```

```python
Title
```

```python
book_id=df[df.title==Title]['bookID'].values[0]
```

```python
book_id
```

```python
scores=list(enumerate(cosine_sim[book_id]))
```

```python
scores
```

```python
sort_scores=sorted(scores,key=lambda x:x[1],reverse=True)
```

```python
sort_scores
```
